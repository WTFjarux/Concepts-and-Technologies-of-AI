Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
plt.figure(figsize=(8, 5))
plt.boxplot(SouthAsia_data['score'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Box Plot of Happiness Score')
plt.xlabel('Score')
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
print("\n\n")

# Box plot for 'Log GDP per capita'
plt.figure(figsize=(8, 5))
plt.boxplot(SouthAsia_data['Log GDP per capita'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Box Plot of Log GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()

```

Here the dots outside the Whiskers are outliners in the data.

```python
# Define outliers using the 1.5 × IQR rule.
# Function to identify outliers using the IQR rule
def find_outliers(series):

    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series[(series < lower_bound) | (series > upper_bound)]

# Identify outliers
outliers_score = find_outliers(SouthAsia_data["score"])
outliers_gdp = find_outliers(SouthAsia_data["Log GDP per capita"])

print("Outliers based on Score:\n")
print(SouthAsia_data[SouthAsia_data["score"].isin(outliers_score)])

print("\nOutliers based on Log GDP per capita:\n")
print(SouthAsia_data[SouthAsia_data["Log GDP per capita"].isin(outliers_gdp)])
```

```python
# Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
outliers = SouthAsia_data[
    SouthAsia_data["score"].isin(outliers_score) |
    SouthAsia_data["Log GDP per capita"].isin(outliers_gdp)
]
non_outliers = SouthAsia_data.drop(outliers.index)


# Plot non-outliers
plt.scatter(non_outliers['Log GDP per capita'], non_outliers['score'],
            color='blue', label='Non-Outliers')

# Plot outliers
plt.scatter(outliers['Log GDP per capita'], outliers['score'],
            color='red', label='Outliers')

# Add labels, title, and legend
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Scatter Plot of GDP per Capita vs. Happiness Score (with Outliers Highlighted)')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.

```python
# Identify outlier countries in both regions based on Score and GDP per Capita.
def identify_outliers(data, column):
    # Calculate Q1, Q3, and IQR
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Identify outliers for South Asia
SA_score_outliers = identify_outliers(SouthAsia_data, 'score')
SA_gdp_outliers = identify_outliers(SouthAsia_data, 'Log GDP per capita')

# Identify outliers for Middle East
ME_score_outliers = identify_outliers(MiddleEastern_data, 'score')
ME_gdp_outliers = identify_outliers(MiddleEastern_data, 'Log GDP per capita')

# Display results
print("South Asia - Score Outliers:\n", SA_score_outliers[['Country name', 'score']])
print("\nSouth Asia - GDP Outliers:\n", SA_gdp_outliers[['Country name', 'Log GDP per capita']])
print("\nMiddle East - Score Outliers:\n", ME_score_outliers[['Country name', 'score']])
print("\nMiddle East - GDP Outliers:\n", ME_gdp_outliers[['Country name', 'Log GDP per capita']])

```
